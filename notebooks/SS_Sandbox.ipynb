### Preliminaries

```python
import sqlite3
import pandas as pd
```

Connect to a database.

```python
conn = sqlite3.connect('db_creation/astronomy.db')
```

Tack a curser to the connection to read the data.

```python
cur = conn.cursor()
```

```python
# Select and display all the data in a pandas dataframe.
query = """
        SELECT *
        FROM solar_system_20;
        """
cur.execute(query)
```

```python
df_ss = pd.DataFrame(cur.fetchall())
df_ss.columns = [x[0] for x in cur.description]  # for column names
df_ss
```
